```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("StudentsPerformance.csv")
print(df.head())
print(df.info())
```

```python
#Histogram
sns.histplot(df['math score'], bins=5, color="purple")
plt.title("Histogram - Math Score")
plt.show()
```

```python
# Box plot
sns.boxplot(x='gender', y='math score', data=df)
plt.title("Boxplot - Math Score by Gender")
plt.show()
```

```python
#Line plot
sns.lineplot(x=range(len(df)), y=df['reading score'], color="green")
plt.title("Line Plot - Reading Scores")
plt.show()
```

```python
#  Bar plot
sns.barplot(x='gender', y='writing score', data=df, color="red")
plt.title("Bar Plot - Writing Score by Gender")
plt.show()
```

```python
#Scatter plot
sns.scatterplot(x='math score', y='writing score', hue='gender', data=df)
plt.title("Scatter Plot - Math vs Writing Score")
plt.show()
```

```python
# Relplot
sns.relplot(x='math score', y='reading score', hue='lunch', data=df)
plt.suptitle("Relplot - Math vs Reading Score by Lunch", y=1.02)
plt.show()
```

```python
# Relation plot
sns.relplot(x="math score", y="reading score", hue="gender", kind="scatter", data=df)
plt.suptitle("Relation Plot - Math vs Reading Score by Gender", y=1.02)
plt.show()
```

```python
# Joint plot
sns.jointplot(x='reading score', y='writing score', data=df, color="purple")
plt.show()
```

```python
# Multivariate (Pair Grid)
g = sns.PairGrid(df[['math score','reading score','writing score','gender']], hue="gender")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.show()
```

```python
# 10. Heatmap
corr = df[['math score','reading score','writing score']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Heatmap - Score Correlations")
plt.show()
```

```python
# Pie chart - Gender distribution
df['gender'].value_counts().plot.pie(
    autopct='%1.1f%%',
)
plt.title("Pie Chart - Gender Distribution")
plt.ylabel("")   #
plt.show()
```
